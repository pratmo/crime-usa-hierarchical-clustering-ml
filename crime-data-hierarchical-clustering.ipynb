# Crime Data Problem

## 1. Hierarchical Clustering

### Import libraries

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
import os
```

```python
os.chdir("/home/prathikm/Desktop/ds_assignments/doing/clustering")
df = pd.read_csv("crime_data.csv") #import universities dataset
df.head()
```

# Re-naming Column

```python
#renaming column 1 to state
df_1 = df.rename({"Unnamed: 0": "State"}, axis =1)
df_1
```

# Data Normalization

```python
#Defining the normalization function
def normalf(i):
    x = (i - i.min())/(i.max()-i.min())
    return(x)
```

```python
#normalize the dataframe
df_normal = normalf(df_1.iloc[:,1:])
```

```python
df_normal
```

# Dendrogram

```python
#Plotting the dendrogram - WARD METHOD
dendrogram = sch.dendrogram(sch.linkage(df_normal,method='ward'))
```

# Fit and Predict

```python
#specifying hierarchical clustering, clusters as 3
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'ward')
```

```python
#prediction - there is no centroid in agglomerative options
y_hc = hc.fit_predict(df_normal)
y_hc
```

# 3 clusters are formed
# We add that as a column in the original dataset

```python
#3 clusters are formed.
#creating new dataframe column as clusters
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
df_1['Hierarchical_clusterid'] = Clusters
```
